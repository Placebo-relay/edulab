# Super Power Data (14)

```python
import pandas as pd
```

```python
PATH = 'super_hero_powers.csv'
original_data = pd.read_csv(PATH)

original_data.head()
```

```python
new_data = original_data.iloc[:, 0].to_frame()
new_data["ability_count"] = original_data.iloc[:, 1:].sum(axis=1)
new_data["hero_abilities"] = original_data.iloc[:, 1:].apply(
    lambda x: ", ".join(x.index[x]), axis=1
)
new_data.head()
```

```python
new_data
```

```python
# @title ability_count 1

from matplotlib import pyplot as plt
new_data['ability_count'].plot(kind='hist', bins=20, title='ability_count')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title ability_count 2

from matplotlib import pyplot as plt
new_data['ability_count'].plot(kind='line', figsize=(8, 4), title='ability_count')
plt.gca().spines[['top', 'right']].set_visible(False)
```
